```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x, y, 'red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Data Contoh')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'b*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
#fig.add_axes(jarak dari kiri, jarak dari kanan, panjang kanvas, lebar kanvas) )

axes.plot(x,y, 'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y, 'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('Y utama')
axes1.set_title('Plot Utama')

axes2.plot(y,x, 'r--')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('Y dalam')
axes2.set_title('Plot Dalam')
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
#figsize=pengukuran pada kanvas yang akan dibuat (width,height), dpi adalah bentuk hasil visual
```

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=100)

axes.plot(x,y,'r')

axes.plot(x,y, 'r*')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('PlotJudul')

fig.savefig("contoh.png",dpi=100)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label='Data x pangkat 2')
axes.plot(x, x**3, label='Data x pangkat 3')

axes.legend(loc=0)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color='red', linewidth=0.25)
axes.plot(x, x+2, color='red', linewidth=0.5)
axes.plot(x, x+3, color='red', linewidth=0.75)
axes.plot(x, x+4, color='red', linewidth=1)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color='green', lw=3, linestyle='-')
axes.plot(x, x+2, color='green', lw=3, ls='-.')
axes.plot(x, x+3, color='green', lw=3, ls=':')
axes.plot(x, x+4, color='green', lw=3, ls='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

line, = axes.plot(x, x+1, color='violet', lw=1.5)
line.set_dashes([5,5,15,5]) 
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color='cyan', lw=3, ls='-', marker='+')
axes.plot(x, x+2, color='cyan', lw=3, ls='-', marker='o')
axes.plot(x, x+3, color='cyan', lw=3, ls='-', marker='s')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color='cyan', lw=3, ls='-', marker='+', markersize='10')
axes.plot(x, x+2, color='cyan', lw=3, ls='-', marker='o', markerfacecolor='blue')
axes.plot(x, x+3, color='cyan', lw=3, ls='-', marker='s', markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```
