```python
left = {
    'key1' : ['K0','K0','K1','K2'],
    'key2' : ['K0','K1','K0','K1'],
    'A' : ['A0','A1','A2','A3'],
    'B' : ['B0','B1','B2','B3'],
}

right = {
    'key1' : ['K0','K1','K1','K2'],
    'key2' : ['K0','K0','K0','K0'],
    'C' : ['C0','C1','C2','C3'],
    'D' : ['D0','D1','D2','D3'],
}

dfLeft = pd.DataFrame(left)
dfRight = pd.DataFrame(right)
```

```python
dfLeft
```

```python
dfRight
```

```python
#Outer data akan ditampilkan semua, ketika ada data yang tidak cocok akan ditulis sebagai data Nan
pd.merge(dfLeft,dfRight,how='outer',on=['key1','key2'])
```

```python
#Inner penggabungan ketika ada data yang tidak cocok/NaN maka tidak akan ditampilkan
pd.merge(dfLeft,dfRight,how='inner',on=['key1','key2'])
```

```python
#Right fokus pada data right
pd.merge(dfLeft,dfRight,how='right',on=['key1','key2'])
```

```python
#Left
pd.merge(dfLeft,dfRight,how='left',on=['key1','key2'])
```
